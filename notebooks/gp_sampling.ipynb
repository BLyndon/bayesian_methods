# Sampling from a Gaussian process

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy.spatial import distance_matrix
```

#### Define the kernel function

```
def kernel(d, l=1):
    return np.exp(-d**2 / (2 * l**2 ))
```

#### Define the covariance

```
def covariance(x, l=1):
    x = np.reshape(x, [-1, 1])
    d = distance_matrix(x, x)
    return kernel(d, l)
```

#### Create the distribution

```
n = 1000
x = 10*np.random.rand(n)

l = 1
mean = np.zeros(n)
cov = covariance(x, l)
```

#### Draw samples from the distribution

```
for i in range(5):
    y = np.random.multivariate_normal(mean=mean, cov=cov)
    plt.scatter(x, y)

plt.show()
```

#### Draw samples from distributions with different correlation lengths

```
for i in range(5):
    l = 0.5*(i+1)
    y = np.random.multivariate_normal(mean=mean, cov=covariance(x, l))
    plt.scatter(x, y, label="l = " + str(l))

plt.legend()
plt.show()
```

```
l = 0.0001
y = np.random.multivariate_normal(mean=mean, cov=covariance(x, l))
plt.scatter(x, y, label="l = " + str(l))
l = 1000
y = np.random.multivariate_normal(mean=mean, cov=covariance(x, l))
plt.scatter(x, y, label="l = " + str(l))

plt.legend()
plt.show()
```
